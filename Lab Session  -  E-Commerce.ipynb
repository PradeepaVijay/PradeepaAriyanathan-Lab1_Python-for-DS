### Q1. Create tables for supplier, customer, category, product, productDetails, order, rating to store the data for the E-commerce with the schema definition given below.


- **`supplier`**(SUPP_ID int primary key, SUPP_NAME varchar(50), SUPP_CITY varchar(50), SUPP_PHONE varchar(10))


- **`customer`** (CUS_ID INT NOT NULL, CUS_NAME VARCHAR(20) NULL DEFAULT NULL, CUS_PHONE VARCHAR(10), CUS_CITY varchar(30) ,CUS_GENDER CHAR,PRIMARY KEY (CUS_ID))


- **`category`** (CAT_ID INT NOT NULL, CAT_NAME VARCHAR(20) NULL DEFAULT NULL,PRIMARY KEY (CAT_ID))


- **`product`** (PRO_ID INT NOT NULL, PRO_NAME VARCHAR(20) NULL DEFAULT NULL, PRO_DESC VARCHAR(60) NULL DEFAULT NULL, CAT_ID INT NOT NULL,PRIMARY KEY (PRO_ID),FOREIGN KEY (CAT_ID) REFERENCES CATEGORY (CAT_ID))


- **`product_details`** (PROD_ID INT NOT NULL, PRO_ID INT NOT NULL, SUPP_ID INT NOT NULL, PROD_PRICE INT NOT NULL,
  PRIMARY KEY (PROD_ID),FOREIGN KEY (PRO_ID) REFERENCES PRODUCT (PRO_ID), FOREIGN KEY (SUPP_ID) REFERENCES SUPPLIER(SUPP_ID))
  
  
- **`order`** (ORD_ID INT NOT NULL, ORD_AMOUNT INT NOT NULL, ORD_DATE DATE, CUS_ID INT NOT NULL, PROD_ID INT NOT NULL,PRIMARY KEY (ORD_ID),FOREIGN KEY (CUS_ID) REFERENCES CUSTOMER(CUS_ID),FOREIGN KEY (PROD_ID) REFERENCES PRODUCT_DETAILS(PROD_ID))


- **`rating`** (RAT_ID INT NOT NULL, CUS_ID INT NOT NULL, SUPP_ID INT NOT NULL, RAT_RATSTARS INT NOT NULL,PRIMARY KEY (RAT_ID),FOREIGN KEY (SUPP_ID) REFERENCES SUPPLIER (SUPP_ID),FOREIGN KEY (CUS_ID) REFERENCES CUSTOMER(CUS_ID))

```python
## Again connect to the Mysql server and while connecting Choose 'ecommerce' database

connection = mysql.connector.connect(host ="localhost",
                                     user ="root",
                                     passwd ="deepavijay",
                                     database = 'ecommerce')
 
## creating a cursor object
cursorObject = connection.cursor()
## Solution for Q1: Creating tables 
table_creation_query = """create table supplier(SUPP_ID int primary key, SUPP_NAME varchar(50), SUPP_CITY varchar(50), SUPP_PHONE varchar(50));
create table customer(CUS_ID INT NOT NULL, CUS_NAME VARCHAR(20) NULL DEFAULT NULL, CUS_PHONE VARCHAR(10), CUS_CITY varchar(30) ,CUS_GENDER CHAR,PRIMARY KEY (CUS_ID));
create table category (CAT_ID INT NOT NULL, CAT_NAME VARCHAR(20) NULL DEFAULT NULL,PRIMARY KEY (CAT_ID));
create table product (PRO_ID INT primary key NOT NULL, PRO_NAME VARCHAR(20) NULL DEFAULT NULL, PRO_DESC VARCHAR(60) NULL DEFAULT NULL,CAT_ID int references category(CAT_ID));
create table product_details(PROD_ID INT primary key NOT NULL, PROD_PRICE INT NOT NULL,PRO_ID int NOT NULL REFERENCES PRODUCT (PRO_ID), SUPP_ID int not null REFERENCES SUPPLIER(SUPP_ID));
create table order_table (ORDID int primary key NOT NULL, ORD_AMOUNT int NOT NULL, ORD_DATE DATE, CUS_ID int NOT NULL REFERENCES CUSTOMER(CUS_ID), PROD_ID int NOT NULL REFERENCES PRODUCT_DETAILS(PROD_ID));
create table rating(RAT_ID INT primary key NOT NULL, CUS_ID INT NOT NULL references CUSTOMER(CUS_ID) , SUPP_ID INT NOT NULL REFERENCES SUPPLIER (SUPP_ID), RAT_RATSTARS INT NOT NULL);"""
# Executing the query
cursorObject.execute(table_creation_query)
connection.close()
```

### Q3) Display the number of the customer group by their genders who have placed any order of amount greater than or equal to Rs.3000.

```python
import mysql.connector
import pandas as pd
try:
    connection = mysql.connector.connect(host ="localhost",
                                     user ="root",
                                     passwd ="deepavijay",
                                     database = 'ecommerce')
    question_one = """SELECT 
                        cus_gender, COUNT(cus_gender)
                    FROM
                        customer c
                            JOIN
                        order_table o ON o.cus_id = c.cus_id
                    WHERE
                        o.ord_amount >= 3000
                    GROUP BY c.cus_gender; """
    result_dataFrame = pd.read_sql(question_one,connection)
    print(result_dataFrame)
    connection.close() #close the connection
except Exception as e:
    connection.close()
    print(str(e))
```

### Q4) Display all the order along with product name ordered by a customer having Customer_Id=2;

```python
### Q4) Display all the order along with product name ordered by a customer having Customer_Id=2;
import mysql.connector
import pandas as pd
try:
    connection = mysql.connector.connect(host ="localhost",
                                     user ="root",
                                     passwd ="deepavijay",
                                     database = 'ecommerce')
    question_two="""SELECT 
                    o.*, pr.pro_name, pr.pro_desc
                FROM
                    order_table o
                        JOIN
                    product_details p ON o.prod_id = p.prod_id
                        JOIN
                    product pr ON pr.pro_id = p.pro_id
                WHERE
                    o.cus_id = 2;"""
    result_dataFrame = pd.read_sql(question_two,connection)
    print(result_dataFrame)
    connection.close() #close the connection
except Exception as e:
    connection.close()
    print(str(e))
```

### Q5) Display the Supplier details who can supply more than one product.

```python
### Q5) Display the Supplier details who can supply more than one product.
import mysql.connector
import pandas as pd
try:
    connection = mysql.connector.connect(host ="localhost",
                                     user ="root",
                                     passwd ="deepavijay",
                                     database = 'ecommerce')
    question_five="""SELECT 
                    s.*
                FROM
                    supplier s
                        JOIN
                    product_details p ON s.supp_id = p.supp_id
                GROUP BY p.supp_id
                HAVING COUNT(p.supp_id) > 1;"""
    result_dataFrame = pd.read_sql(question_five,connection)
    print(result_dataFrame)
    connection.close() #close the connection
except Exception as e:
    connection.close()
    print(str(e))
```

### Q6) Find the category of the product whose order amount is minimum.

```python
# 6. Find the category of the product whose order 
#amount is minimum

import mysql.connector
import pandas as pd
try:
    connection = mysql.connector.connect(host ="localhost",
                                     user ="root",
                                     passwd ="deepavijay",
                                     database = 'ecommerce')
    question_six="""SELECT 
                        c.*, o.ord_amount
                    FROM
                    order_table o
                        JOIN
                    product_details pd ON o.prod_id = pd.prod_id
                        JOIN
                    product p ON p.pro_id = pd.pro_id
                        JOIN
                    category c ON c.cat_id = p.cat_id
                        HAVING MIN(o.ord_amount);"""
    result_dataFrame = pd.read_sql(question_six,connection)
    print(result_dataFrame)
    connection.close() #close the connection
except Exception as e:
    connection.close()
    print(str(e))
```

### Q8) Print the top 3 supplier name and id and rating on the basis of their rating along with the customer name who has given the rating.

```python
# 8. Print the top 3 supplier and id and their rating
# on the basis of their rating alon with the 
# customer name who has given the rating 
import pandas as pd
try:
    connection = mysql.connector.connect(host ="localhost",
                                     user ="root",
                                     passwd ="deepavijay",
                                     database = 'ecommerce')
    question_eight="""SELECT 
                    s.supp_id, s.supp_name, r.rat_ratstars, c.cus_name
                FROM
                    supplier s
                        JOIN
                    rating r ON s.supp_id = r.supp_id
                        JOIN
                    customer c ON r.cus_id = c.cus_id
                ORDER BY r.rat_ratstars DESC
                LIMIT 3;"""
    result_dataFrame = pd.read_sql(question_eight,connection)
    print(result_dataFrame)
    connection.close() #close the connection
except Exception as e:
    connection.close()
    print(str(e))
```
